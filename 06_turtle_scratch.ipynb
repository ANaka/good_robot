```python
xr, yr = np.meshgrid(np.arange(5), np.arange(5))

turtles = []
for x, y in zip(xr.flatten(), yr.flatten()):
    turtles.append(Turtle(x=x, y=y))
```

```python
for tur in turtles:
    tur.pen_down()
    tur.forward(0.1)
    tur.circle(radius=0.1, extent=90, use_degrees=True)
    tur.forward(0.1)
    tur.circle(radius=0.1, extent=90, use_degrees=True)
    tur.forward(0.1)
    tur.circle(radius=0.1, extent=90, use_degrees=True)
```

```python
all_lines = np.concatenate([tur.lines for tur in turtles], axis=1)
```

```python
lc = LineCollection(layer_to_lines(all_lines))
```

```python
f,ax = plt.subplots()
ax.add_collection(lc)
ax.axis('tight')
ax.axis('square')
```
